### 4. Comibining Customers with the other DF

#### Importing the other df

```python
ords_prods_merge = pd.read_pickle(os.path.join(path, 'Prepared Data', 'orders_products_with_flags_v2.pkl'))
```

```python
# comlpeting the missing cleaning step from the exercise, so that ords_prods_merge is clean and accurate
ords_prods_merge.loc[ords_prods_merge['prices'] >100, 'prices'] = np.nan
```

```python
# checking result 
ords_prods_merge['prices'].max()
```

```python
# checking ords_prods_merge df and datatypes
ords_prods_merge.info()
```

#### user_id, which is our key is int64 in both df

```python
ords_prods_merge['user_id'].describe()
```

#### ords_prods_merge has the same max for user_id as customers this means that user_id is a fully matching column

```python
#removing _merge column in preparation for merge
ords_prods_merge=ords_prods_merge.drop(columns=['_merge'])
```

```python
ords_prods_cust_merge = pd.merge(ords_prods_merge, customers, on = 'user_id', indicator = True)
```

```python
#Confirm the results of the merge using the merge flag
ords_prods_cust_merge.head()
```

```python
#Confirm the results of the merge using the merge flag with frequency
ords_prods_cust_merge['_merge'].value_counts()
```

#### Merge worked based on head syntax and the merge flag, because the ords_prods_merge originally had  32435059 rows.

```python
# _merge flag isn't needed anymore
ords_prods_cust_merge=ords_prods_cust_merge.drop(columns=['_merge'])
```
