### 3. Comparing Customer Behaviour in Different Geographic Areas

#### Need to create a “Region” column based on the “State” column -> similar to creating flags, so new variables, will do it with if For-Loops. 

#### First, I want to check if all 51 states appear in the STATE column or only certain ones

```python
#Value_counts fot STATE columng
ords_prods_cust['STATE'].value_counts().sort_index(ascending=True)
```

#### Since it was faster to copy it to a spreadsheet than finding the syntax to assign index number to the states or calculating the number of different values the column has, I confirm that all 51 states appear in the df.

```python
# for loop to create region values
result_region = []

for value in ords_prods_cust["STATE"]:
 if value == 'Maine' or value == 'New Hampshire' or value == 'Vermont' or value =='Massachusetts' or value =='Rhode Island' or value =='Connecticut' or value =='New York' or value =='Pennsylvania' or value =='New Jersey':
    result_region.append("Northeast")
 elif value == 'Wisconsin' or value =='Michigan' or value =='Illinois' or value =='Indiana' or value =='Ohio' or value =='North Dakota' or value =='South Dakota' or value =='Nebraska' or value =='Kansas' or value =='Minnesota'or value =='Iowa' or value =='Missouri':
    result_region.append("Midwest")
 elif value == 'Delaware' or value =='Maryland' or value =='District of Columbia' or value =='Virginia' or value =='West Virginia' or value =='North Carolina' or value =='South Carolina' or value =='Georgia' or value =='Florida' or value =='Kentucky'or value =='Tennessee' or value =='Mississippi'or value =='Alabama' or value =='Oklahoma' or value =='Texas' or value =='Arkansas' or value =='Louisiana':
    result_region.append("South")
 else:
    result_region.append("West")
```

```python
# checking new list of regions by printing result_region
result_region
```

```python
#making result_region list into a new column of DF
ords_prods_cust["REGION"]=result_region
```

```python
#checking result of new region columns with value_counts
ords_prods_cust["REGION"].value_counts(dropna=False)
```

#### Determining whether there’s a difference in spending habits between the different U.S. regions

```python
# creating a crosstab with region and spending_flag with pd.crosstab
crosstab = pd.crosstab(ords_prods_cust['REGION'], ords_prods_cust['spending_flag'], dropna = False)
```

```python
crosstab
```

#### Copying this into a spreadshit a calculating the % of the values, it is clear that in all regions only around 2% of the customers are high spender, the rest is low spender

### 5. Customer Profiling

#### Profiling variable based on age -> will create a flag with age groups with for loops

```python
# check min age to determine smallest age group
ords_prods_cust['Age'].describe()
```

```python
#creating age_group list with forloops
result_age = []

for value in ords_prods_cust["Age"]:
 if value <= 24:
    result_age.append('18-24')
 elif value >= 25 and value <35:
    result_age.append('25-34')
 elif value >= 35 and value <44:
    result_age.append('35-44')
 elif value >= 45 and value <54:
    result_age.append('45-54')
 elif value >= 55 and value <64:
    result_age.append('55-64')
 elif value >= 65 and value <74:
    result_age.append('65-74')
 else:
    result_age.append('75-')
```

```python
result_age
```

```python
#assigning age group list as age group column
ords_prods_cust['age_group']=result_age
```

```python
#checking with head function
ords_prods_cust.head()
```

```python
#checking new age group colulm with value counts
ords_prods_cust['age_group'].value_counts(dropna=False)
```

#### Is is unexpected that most of the purchases are done by people above 75 years old

#### Profiling variable based on income -> will create a flag with income groups with for loops

```python
# check income stat to determine smallest income group
ords_prods_cust['income'].describe()
```

```python
#creating income group list with forloops
result_income = []

for value in ords_prods_cust['income']:
 if value <= 32048:
    result_income.append('Poor or near_poor')
 elif value >= 32049 and value <53413:
    result_income.append('Lower-middle class')
 elif value >= 53414 and value <106827:
    result_income.append('Middle class')
 elif value >= 106828 and value <373894:
    result_income.append('Upper-middle class')
 else:
    result_income.append('Rich')
```

```python
result_income
```

```python
#assign income list as income group column
ords_prods_cust['income group']=result_income
```

```python
#checking with head function
ords_prods_cust.head()
```

```python
#checking income group column with value_counts
ords_prods_cust['income group'].value_counts(dropna=False)
```

#### This picture looks more as I expected, more people in the higher income group can afford to buy from an online super market
