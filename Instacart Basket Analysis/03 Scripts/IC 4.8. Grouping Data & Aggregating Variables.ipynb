### 3. Creating Loyalty Flag

```python
# creating max order column
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
pd.options.display.max_rows = None
```

```python
# checking result
ords_prods_merge.head(100)
```

#### Deriving Columns with loc()

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

### 4. Analyzing Spending Habits with Loyalty_Flag

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

#### It seems that the longer a customer knows been shopping with us, the cheaper products they buy.

### 5. Creating Spending Flag

```python
ords_prods_merge['avg_price_per_user'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(100)
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price_per_user'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['avg_price_per_user'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.head(100)
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

### 6. Frequent Customers Flag

```python
ords_prods_merge['median_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(100)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_prior_order'] > 20, 'frequent_customer_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_prior_order'] <= 20) & (ords_prods_merge['max_order'] > 10), 'frequent_customer_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_prior_order'] <= 10, 'frequent_customer_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.head(100)
```

```python
ords_prods_merge['frequent_customer_flag'].value_counts(dropna = False)
```
